```python
# skill -->
import pandas as pd 
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt 

# sample data 

data = {
    'Customer' : ['riya' ,'amen' , 'pisyh' ],
    'Age': [20, 30 , 40 ],
    'Spending': [ 100 , 200, 400]
} 

df = pd.DataFrame(data)

x = df[["Age" , "Spending"  ]]

model = KMeans(n_clusters=2 , random_state=42 , n_init=10)

df['Group'] = model.fit_predict(x)

plt.figure(figsize =(6, 5))
for group in df['Group'].unique():   # [0 , 1]
        group_data = df[df['Group'] == group]
        plt.scatter(group_data['Age'] , group_data["Spending"] , label = f"Group {group}")


```

```python
#  1 pca -->  only important data sent "high and low amnd fig" , overfiting 

# 10 clusing 1 katurl and pca -->  less of inport column and figuree 

# model tring in very slow 
#  visulire --> fix
# overfitting --> [] , [] ,[] , [ ] , [] , [] ,[] , [] , [] ,[] ,[]

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA 


data = {
    'Age':[24, 34, 34, 34, ],
    'Income':[9000000, 3300000998,900999090 ,9932938039],
    'Spending': [70, 90 , 90 , 90],
    'Saving': [1000 , 5000 , 4000, 1000]
}


#  this is  clean  
df = pd.DataFrame(data)

scaler = StandardScaler()

scaled_data = scaler.fit_transform(df)  # only fit aaareg used in data


pca = PCA(n_components = 2)
pca_result = pca.fit_transform(scaled_data)

pca_df = pd.DataFrame(pca_result , columns = ['PCA1' , 'PCA2'])

explained_variance = pca.explained_variance_ratio_ 
print("varince ")
print(np.round(explained_variance * 100 , 2)) # 5.56  0.56 0.50 
plt.figure(figsize=(8,7))
plt.scatter(pca_df['PCA1'] , pca_df['PCA2'] , color = "black" , s=80)
plt.xlabel('PCA1 main Pattern')
plt.ylabel('PCA2 minor Pattern')
plt.grid(True)
plt.show()

print("New  data with ")
print(pca_df)
```
